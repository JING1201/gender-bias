```python
import nltk as nl
```

```python
import re
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
letter_dir = '/Users/vassiki/Desktop/gender/example_letters/'
letter_m_fn = os.path.join(letter_dir,'letterofRecM')
letter_f_fn = os.path.join(letter_dir,'letterofRec_W')
```

```python
def create_tokens(fn):
    with open(fn) as f:
        raw = f.read()
    raw = re.sub(r'\W+', ' ', raw)
    tokens = nl.word_tokenize(raw)
    return(tokens)    
```

```python
tokens_m = create_tokens(letter_m_fn)
tokens_f = create_tokens(letter_f_fn)
```

```python
def add_tags(tokens):
    tags = nl.pos_tag(tokens)
    tags_nn = [w for idx,w in enumerate(tags) if tags[idx][1] == "NN"]
    pos = ["NN","JJS","JJR","JJ","VB"]
    counts = []
    for pos_t in pos:
        tags_pos = [w for idx,w in enumerate(tags) if tags[idx][1] == pos_t]
        counts.append(len(tags_pos))
    
    counts = [float(c)/len(tokens) for c in counts]
    print counts
    x_marks = np.arange(len(pos))
    plt.bar(x_marks, counts, align="center")
    plt.xticks(x_marks, pos)
    plt.show()
```

```python
add_tags(tokens_f)
```
